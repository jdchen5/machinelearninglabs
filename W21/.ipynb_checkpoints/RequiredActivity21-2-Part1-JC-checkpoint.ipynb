```python
import numpy as np
from scipy import signal

sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)

noise_level = 0.25 # a number between 0 and 1

sig_noise = sig + noise_level * np.random.randn(len(sig))

sq_wave = np.ones(128)

corr = signal.correlate(sig_noise, sq_wave, mode='same') / 128
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(sig, label='Original Signal')
plt.plot(sig_noise, label='Noisy Signal')
plt.plot(corr, label='Correlated Signal')
plt.legend()
plt.show()
```

```python
from scipy.linalg import toeplitz

a = [1., 0.7, -0.3]

b = [0., 1., 1., 0., 1., 1., 0.]

len_b = len(b)
print(len_b)

# code to create Toeplitz matrices

first_row = np.r_[np.array([a[0], 0., 0.]), np.zeros(len_b - 3)]
print(first_row)

first_col = np.r_[np.array(a), len_b - 1]
print(first_col)

X_filter = toeplitz(first_col, first_row)

print(X_filter)
```

```python
from scipy.sparse.linalg import lsqr # import the least squares module

k = 3  # Length of vector 'x'
n = len(a)  # Length of vector 'a'

e1 = np.zeros(k + n - 1)

e1[0] = 1

# Construct the first column of the Toeplitz matrix, which is just vector 'a' followed by zeros
first_col = np.r_[a, np.zeros(n - 1)]

first_row = np.zeros(n+k - 1)
first_row[0] = a[0]

M_a = toeplitz(first_col, first_row)  # Toeplitz matrix representation of a * _

print(M_a.shape)
print(e1.shape)

x = lsqr(M_a, e1)[0] # solve the equation M * x = e_1

x
```

```python
from scipy.sparse.linalg import lsqr # import the least squares module

k = 7  # Length of vector 'x'
n = len(a)  # Length of vector 'a'

e1 = np.zeros(k + n - 1)
print(f"e1's shape = {e1.shape}")

e1[0] = 1

# Construct the first column of the Toeplitz matrix, which is just vector 'a' followed by zeros
#The first column should be of length k + n - 1 because it represents the result of convolving a with x, where x has yet to be determined.
first_col = np.r_[a, np.zeros(k - 1)]

# Construct the first row of the Toeplitz matrix, which is the first element of 'a' followed by zeros
first_row = np.zeros(k+n - 1)
first_row[0] = a[0]

M_a = toeplitz(first_col, first_row)  # Toeplitz matrix representation of a * _

print(f"M_a's shape = {M_a.shape}")


x = lsqr(M_a, e1)[0] # solve the equation M * x = e_1

x
```
