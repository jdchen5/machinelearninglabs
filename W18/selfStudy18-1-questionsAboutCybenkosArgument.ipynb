```python
# if you're using jupyter
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def sigmoid(z):
    """The sigmoid function."""
    return 1.0/(1.0+np.exp(-z))

x = np.arange(-10., 10., .1)

plt.plot(x, sigmoid(x))
```

```python
# example of python closure 
def shape_shift(f, w=1., b=0.):
    g = lambda x: f(w*x + b)
    return np.vectorize(g)

# to add 2 or more functions in python 
# for function superposition
def add_functions(fs, ws):   #fs: list of functions; ws: list of weights
    assert len(fs) == len(ws) 
    _sum = lambda x: sum([w*f(x) for f, w in zip(fs, ws)])
    return _sum
```

```python
w,b=20.,0.
shifted_sigmoid_1=shape_shift(sigmoid,w,b)
w,b=20.,-20.
shifted_sigmoid_2=shape_shift(sigmoid,w,b)

# Plot the original sigmoid function
plt.plot(x, sigmoid(x), label='Sigmoid',color='black')

plt.plot(x, shifted_sigmoid_1(x), label = 'Shifted Sigmoid 1', color='green')
plt.plot(x, shifted_sigmoid_2(x), label = 'Shifted Sigmoid 2', color='orange')
```

```python
weight_1, weight_2=0.8, -0.8

# Define the combined function
shifted_sigmoid_3 = add_functions([shifted_sigmoid_1, shifted_sigmoid_2], [weight_1, weight_2])


# Plot the original sigmoid function
plt.plot(x, sigmoid(x), label='Sigmoid',color='black')

plt.plot(x, shifted_sigmoid_1(x), label = 'Shifted Sigmoid 1', color='green')
plt.plot(x, shifted_sigmoid_2(x), label = 'Shifted Sigmoid 2', color='orange')
plt.plot(x, shifted_sigmoid_3(x), label = 'Shifted Sigmoid 3', color='red')
```
