# Guide and Ideas for Capstone Project - Function 1

# Searching for Contamination Sources
This may sound simple because you only have a two-dimensional input, however it is a very difficult problem. It corresponds to trying to find the source of radiation in some square area. However, you can only detect the radiation once you are very close to it, meaning most of the readings will be zero. There are two sources, one is not too dangerous, so make sure you try to find both modes of the function.

We begin this guide by downloading the data:

```python
X = np.load('initial_data/function_1/initial_inputs.npy')
Y = np.load('initial_data/function_1/initial_outputs.npy')
```

```python
def format_query(point, NoOfDigits):
    return '-'.join(f"{x:.{NoOfDigits}f}" for x in point)
```

```python
# Print X with 16 digits of precision
for row in X:
    print(format_query(row,16))

# Print a separator line
print("-" * 50)

# Print Y with 16 digits of precision
print(format_query(Y,16))
```

## Random Search

The simplest solution would be a simple random search, that is, we can randomly choose our next query point:

```python
next_query = np.random.uniform(size = 2)
print(format_query(next_query,6))
```

While this solution is easy to implement, we know it will be very slow. However, it could serve as a placeholder for gathering more information while you research which method you want to use for each function.

```python
# Define the dimension and the number of samples
dim = 2  # Two-dimensional input space
n_samples = 1  # Number of initial samples

# Initialize Latin Hypercube Sampler
lhs = LatinHypercube(d=dim, seed=0)

# Generate sample points
initial_samples = lhs.random(n=n_samples)

# Scale samples to your required bounds, e.g., [0, 1] for each dimension
initial_samples = initial_samples  # Assuming bounds are [0, 1]

# Assuming bounds are [0, 1] for both dimensions
# No additional scaling needed if your bounds are already [0, 1]

# Format the output for each point
formatted_output = [format_query(point,6) for point in initial_samples]

# Print each formatted point
for point in formatted_output:
    print(point)
```

# Upper Confidence Bound

A second alternative would be to use Bayesian Optimization and consider an Upper Confidence Bound acquisition function:

```python
# fits a Gaussian Process Regressor model to data X (input features) and Y (output/target values)
gpr = GaussianProcessRegressor()
gpr.fit(X, Y)

# to optimize the acquisition function, we will simply use gridsearch over a space of 10.000 gridpoints
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)

#Creating a Grid for Optimisation
X_grid = []
for i in range(len(x1)):
    for j in range(len(x2)):
        X_grid.append([x1[i], x2[j]])

X_grid = np.array(X_grid)

#Upper Confidence Bound (UCB) Acquisition Function
mean, std = gpr.predict(X_grid, return_std = True)
#(1.96 in this case, corresponding to the 95% confidence interval in a normal distribution)
ucb = mean + 1.96 * std

#Selecting the Next Query Point
idx_max = np.argmax(ucb)
next_query = X_grid[idx_max]
print(next_query)
```

Solve the UCB in an area restricted around

Below we provide code for both:

```python
#Appoach 1

# Random search around an area restricted around $x$ = [0.650114059798764, 0.68152635206021]
next_query_x1 = np.random.normal(0.650114059798764, scale = 0.01)
next_query_x2 = np.random.normal(0.68152635206021, scale = 0.01)
print('Next query Random: ', [next_query_x1, next_query_x2])


#Approach 2
# Using a restricted UCB
gpr = GaussianProcessRegressor()
gpr.fit(X, Y)

# to optimize the acquisition function, we will simply use gridsearch over a space of 10.000 gridpoints
#the NumPy library's linspace function to generate an array x1 of 100 evenly spaced numbers: start of 0.6; end of 0.65; 100 samples to generate
x1 = np.linspace(0.6, 0.65, 100)
x2 = np.linspace(0.65, 0.69, 100)

X_grid = []
for i in range(len(x1)):
    for j in range(len(x2)):
        X_grid.append([x1[i], x2[j]])

X_grid = np.array(X_grid)
mean, std = gpr.predict(X_grid, return_std = True)
ucb = mean + 1.96 * std

idx_max = np.argmax(ucb)
next_query = X_grid[idx_max]
print('Next query UCB: ', format_query(next_query,6))
```

```python
#Approach 3


# Define the UCB acquisition function, to balance exploration (searching uncertain regions) and exploitation (optimizing in known good regions).
def upper_confidence_bound(X, X_sample, Y_sample, gpr, kappa):
    mu, sigma = gpr.predict(X, return_std=True)
    return mu + kappa * sigma

# Define the function to generate a single candidate point for evaluation
def generate_point_uniformly(bounds):
    return np.random.uniform(bounds[:, 0], bounds[:, 1], bounds.shape[0])


# Placeholder function to submit your query and receive the output
def submit_query_and_receive_output(X):
    # This function should submit the query X and return the received output
    # Implement the submission mechanism here
    # For now, it returns a random number as a placeholder
    return np.random.rand()  # Replace with actual submission and receiving mechanism


# Initialize your Gaussian process with Matern kernel
kernel = Matern(length_scale=1.0) + WhiteKernel(noise_level=1)
gpr = GaussianProcessRegressor(kernel=kernel)

# Initial dataset
X_sample = X
Y_sample = Y

# Kappa parameters for exploration-exploitation balance
#As the loop progresses, kappa decreases, gradually shifting the focus from exploring uncertain regions to exploiting regions known to yield better results.
kappa_initial = 5.0
kappa_decay = 0.9
min_kappa = 0.1


# Set the bounds for each dimension of your input space
bounds = np.array([[0, 1]] * 2)  # Adjust according to your problem's constraints

# Define the number of weeks you plan to run this optimization
N_WEEKS = 10  # adjust this based on your requirement

# Weekly Optimization Loop
for week in range(N_WEEKS):

    # Decay kappa value exponentially
    kappa = max(kappa_initial * (kappa_decay ** week), min_kappa)
    
    # Update Gaussian process with existing samples
    gpr.fit(X_sample, Y_sample)

    # Generate a single point to predict the UCB over
    X_next = generate_point_uniformly(bounds)

    # Select the point based on UCB using updated kappa
    UCB_value = upper_confidence_bound(X_next.reshape(1, -1), X_sample, Y_sample, gpr, kappa)
    Y_next = submit_query_and_receive_output(X_next)

    # Add the new sample to our existing samples
    X_sample = np.vstack((X_sample, X_next))
    Y_sample = np.append(Y_sample, Y_next)

    # print or log the submitted query
    formatted_query = format_query(X_next,6)
    print(f"Week {week + 1}, Submitted Query: {formatted_query}")
```

```python

# Define the UCB acquisition function with fixed kappa
def upper_confidence_bound(gp, X, kappa=2.576):
    mean, std = gp.predict(X, return_std=True)
    return mean + kappa * std

# Function to propose the next sampling point
def propose_next_sample(gp, bounds, n_restarts=10):
    #These will store the best acquisition function value found and the corresponding input point.
    best_value = float('-inf') #is set to negative infinity
    best_x = None
    
    for starting_point in np.random.uniform(bounds[:, 0], bounds[:, 1], size=(n_restarts, dim)):
        res = minimize(lambda x: -upper_confidence_bound(gp, x.reshape(1, -1)), 
                       x0=starting_point, 
                       bounds=bounds)
        if res.fun < best_value:
            best_value = res.fun
            best_x = res.x
            
    return best_x

# Define a Matern kernel
kernel = Matern(nu=1.5)
# Initialize Gaussian Process Regressor
gp = GaussianProcessRegressor(kernel=kernel)

# Fit GP to initial data (assuming you have initial Y values)
gp.fit(X, Y)


# Define the bounds of your search space
bounds = np.array([[0, 1], [0, 1]])  # Assuming [0, 1] for both dimensions



next_sample = propose_next_sample(gp, bounds)

if next_sample is None:
    print("Error: next_sample is None. Check the function that generates next_sample.")
else:
    formatted_query = format_query(next_sample)
    print(formatted_query)
```

Doing this every week, we expect to be able to find the first of the two modes! This should would be enough to provide a good solution to the problem. To really verify if we found the maximum, we would need to find a second mode, however, we may be unlucky and simply never find it. This is an important part of Machine Learning: in theory we might want optimal, perfect solutions, but in practice most of the advancements can be done with simple solutions and careful consideration of the data available to us.

```python
# Validation for Week 1 with optimum_point = np.array([0.6479752188646964, 0.6753847431236562])
def sample_around_optimum(optimum_point, radius=0.05, num_samples=30):
    # Generate random points around the optimum within a given radius
    np.random.seed(0)  # For reproducibility
    delta = np.random.uniform(-radius, radius, size=(num_samples, 2))
    sampled_points = optimum_point + delta
    return sampled_points

optimum_point = np.array([0.6479752188646964, 0.6753847431236562])  # Example optimum point
sampled_points = sample_around_optimum(optimum_point)
```

```python
# Evaluate Samples Using the Model
# Predict values for the sampled points
predicted_values = gpr.predict(sampled_points)
```

```python
# Apply a Statistical Test see if the values at the optimum point are significantly different from the values at the sampled points. t-test: p-value should <0.05
from scipy.stats import ttest_1samp

# Value at the optimum (assuming this is already known)
optimum_value = gpr.predict(optimum_point.reshape(1, -1))

# Perform a one-sample t-test
t_statistic, p_value = ttest_1samp(predicted_values, optimum_value)

print("T-statistic:", t_statistic)
print("p-value:", p_value)
```
